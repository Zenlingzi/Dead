```python
import requests

from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.stockbiz.vn/Investment.aspx")

soup = BeautifulSoup(response.content, "html.parser")

print(soup)
```

```python
titles = soup.findAll('h3', class_='module')

print(titles)
```
